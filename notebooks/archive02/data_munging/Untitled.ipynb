```python
from pathlib import Path
from arcgis.features import GeoAccessor
```

```python
pth_data = Path('../../data')

pth_raw = pth_data/'raw'
pth_raw_gdb = pth_raw/'raw.gdb'

pth_int = pth_data/'interim'
pth_int_gdb = pth_int/'interim.gdb'

pth_coffee = pth_raw_gdb/'coffee'
pth_coffee_int = pth_int_gdb/'coffee'
count = 'COUNT'
name_class = 'NAME_CLASS'
```

```python
df = GeoAccessor.from_featureclass((pth_raw_gdb/'coffee').as_posix())
df = df[['LOCNUM', 'CONAME', 'STREET', 'CITY', 'STATE', 'ZIP', 'ZIP4', 
         'NAICS', 'SIC', 'SHAPE']].copy()
df.head()
```

```python
df_cnt = df[[[c for c in df.columns if c != 'CONAME'][0], 'CONAME']].groupby('CONAME').count()
df_cnt.columns = [count]
df_cnt.sort_values(count, ascending=False, inplace=True)
df_cnt.head()
```

```python
df_cnt[name_class] = df_cnt.index
df_cnt.head(25)
```

```python
df_cnt[name_class] = df_cnt.apply(lambda r: 'INDEPENDENT' if r[count] == 1 else r[name_class], axis=1)
df_cnt.head(25)
```

```python
df_out = df.join(df_cnt, on='CONAME')
df_out.head(10)
```

```python
df_out.spatial.to_featureclass(pth_coffee_int.as_posix())
```

```python
df_out[df_out['NAME_CLASS'] == 'INDEPENDENT'].spatial.to_featureclass((pth_raw_gdb/'coffee_independent').as_posix())
```

```python
df_out[df_out['NAME_CLASS'] != 'INDEPENDENT'].spatial.to_featureclass((pth_raw_gdb/'coffee_competition').as_posix())
```
